```python
mileage_polygons = df[df.leader == True].groupby('polygon')[['mileage_list', 'mileage_telematics']].sum()
```

```python
def calculate_mileage_score(df: pd.DataFrame):
	df = df.copy()
	df['mileage_deviation'] = (1 - (df['mileage_list'] / df['mileage_telematics'])).abs()*100


	return df
```
